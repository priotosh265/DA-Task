```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file1 = "Dump1.xlsx"
file2 = "Dump2.xlsx"
file3 = "Dump3.xlsx"
```

```python
# Read the datasets
data1 = pd.read_excel(file1)
data2 = pd.read_excel(file2)
data3 = pd.read_excel(file3)
```

```python
# Combine the datasets
data = pd.concat([data1, data2, data3], ignore_index=True)
```

```python
# Data Cleaning
# Convert date columns to datetime
data['Instance Date'] = pd.to_datetime(data['Instance Date'], errors='coerce')
data['Last Lead Activity Date'] = pd.to_datetime(data['Last Lead Activity Date'], errors='coerce')
```

```python
# Replace missing or empty values in important columns
data.fillna('Not Available', inplace=True)
```

```python
# Analyze Lead Stage
lead_stage_counts = data['Lead Stage'].value_counts()
print("Lead Stage Distribution:\n", lead_stage_counts)
```

```python
# Analyze Lead Status
lead_status_counts = data['Lead Status'].value_counts()
print("\nLead Status Distribution:\n", lead_status_counts)
```

```python
# Group by Course to find popular courses
popular_courses = data['Course'].value_counts()
print("\nPopular Courses:\n", popular_courses)
```

```python
# Analyze conversion rate (Paid Applications / Total Leads)
conversion_rate = (data['Enrolment Status'].str.lower() == 'yes').sum() / len(data) * 100
print(f"\nConversion Rate: {conversion_rate:.2f}%")
```

```python
# Visualization
# Plot Lead Stage Distribution
plt.figure(figsize=(10, 6))
lead_stage_counts.plot(kind='bar', color='skyblue', title='Lead Stage Distribution')
plt.xlabel('Lead Stage')
plt.ylabel('Number of Leads')
plt.show()
```

```python
# Plot Lead Status Distribution
plt.figure(figsize=(10, 6))
lead_status_counts.plot(kind='bar', color='orange', title='Lead Status Distribution')
plt.xlabel('Lead Status')
plt.ylabel('Number of Leads')
plt.show()
```

```python
# Plot Popular Courses
plt.figure(figsize=(12, 8))
popular_courses.head(10).plot(kind='bar', color='green', title='Top 10 Popular Courses')
plt.xlabel('Course')
plt.ylabel('Number of Enrollments')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
lead_stage = [
    'Untouched', 'Cold', 'No Response', 'NOT INTERESTED', 
    'Warm', 'Invalid', 'Application Submitted', 
    'NOT ELIGIBLE', 'Hot', 'Reject', 'Selected'
]
counts = [74285, 31503, 27020, 9255, 3881, 2534, 619, 581, 245, 74, 3]

# Set Seaborn style
sns.set_theme(style="ticks")

# Create a color palette
colors = sns.color_palette("husl", len(lead_stage))

# Create the bar plot
plt.figure(figsize=(12, 8))
bars = plt.bar(lead_stage, counts, color=colors, edgecolor="white")

# Add value annotations on top of the bars
for bar, count in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 500, 
             f'{count}', ha='center', fontsize=10, color='black')

# Add titles and labels
plt.title("Lead Stage Distribution", fontsize=16, fontweight="bold", color="black")
plt.xlabel("Lead Stage", fontsize=12, fontweight="bold", color="black")
plt.ylabel("Count", fontsize=12, fontweight="bold", color="black")
plt.xticks(rotation=45, fontsize=10, color="black")
plt.yticks(fontsize=10, color="black")

# Add a background grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
# plt.tight_layout()
plt.show()
```
